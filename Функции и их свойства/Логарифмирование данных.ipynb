```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Настройка размеров графика по умолчанию для корректного вывода.
plt.rc('figure', figsize=(12, 8))
```

```python
# Считывание данных. В результате data — это NumPy-массив.
data = pd.read_csv('https://code.s3.yandex.net/Math/datasets/sleep.txt', sep='\t')
data = data[['Species', 'BodyWt', 'BrainWt', 'Danger']].dropna().values
```

```python
data[:, -1]
```

```python
# Настройка размеров графика по умолчанию для корректного вывода.
plt.rc('figure', figsize=(12, 8))

plot = plt.scatter(np.log(data[:, 1].astype('float')), np.log(data[:, 2].astype('float')), c=data[:, 3].astype('float'))

handles, labels = plot.legend_elements()
plt.legend(handles, labels, title='Новый маркетолог')

plt.xlabel('Логарифм массы тела')
plt.ylabel('Логарифм массы мозга')
plt.title('Нужно ли увеличивать мозг?')
plt.show()
```

```python
plot = plt.scatter(data[:, 1], data[:, 2], c=data[:, -1])
plt.xscale('log')
plt.yscale('log')

handles, labels = plot.legend_elements()
plt.legend(handles, labels, title='Новый маркетолог')

plt.xlabel('Логарифм массы тела')
plt.ylabel('Логарифм массы мозга')
plt.title('Нужно ли увеличивать мозг?')
plt.show()
```
