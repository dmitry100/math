```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Создание функции.
def f(t):
	return 50 - 20 * 3**(-t)

# Создание точек для построения графика.
x = np.arange(0, 20, 0.1)
y = f(x)

# Построение графика.
plt.plot(x, y)
```
