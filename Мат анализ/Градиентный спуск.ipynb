**Градиент** - вектор частных производных 

```python
import numpy as np

x = np.array([1.5, 1.5])


#функция
def f(x):
    return (
        (x[0] - 1) ** 2 * (x[1] + 1) ** 4
        + (x[0] + 1) ** 4 * (x[1] - 1) ** 2
        + (x[0] - 3) ** 2 * (x[1] - 2) ** 2
    )


#частная производная по x
def dx(x):
    return (
        2 * (x[0] - 1) * (x[1] + 1) ** 4
        + 4 * (x[0] + 1) ** 3 * (x[1] - 1) ** 2
        + 2 * (x[0] - 3) * (x[1] - 2) ** 2
    )

#частная производная по y

def dy(x):
    return (
        4 * (x[0] - 1) ** 2 * (x[1] + 1) ** 3
        + 2 * (x[0] + 1) ** 4 * (x[1] - 1)
        + 2 * (x[0] - 3) ** 2 * (x[1] - 2)
    )


#значение градиента (вектор частных производных)
def grad(x):
    
    return np.array([dx(x), dy(x)])

#новая точка = разница старой и градиента
result = x - grad(x)

#результат новой точки и значения функции в ней
print(result)
print(f(result))
```

```python
import numpy as np

x = np.array([1.5, 1.5])

# Функция для вывода результатов одного шага
def one_step_print(step_no, x, f, grad):
    print("Шаг " + str(step_no) + ":")
    print("x       = ", x)
    print("f(x)    = ", f)
    print("grad(x) = ", grad)
    print()


def f(x):
    return (
            (x[0] - 1) ** 2 * (x[1] + 1) ** 4
            + (x[0] + 1) ** 4 * (x[1] - 1) ** 2
            + (x[0] - 3) ** 2 * (x[1] - 2) ** 2
    )


def dx(x):
    return (
            2 * (x[0] - 1) * (x[1] + 1) ** 4
            + 4 * (x[0] + 1) ** 3 * (x[1] - 1) ** 2
            + 2 * (x[0] - 3) * (x[1] - 2) ** 2
    )


def dy(x):
    return (
            4 * (x[0] - 1) ** 2 * (x[1] + 1) ** 3
            + 2 * (x[0] + 1) ** 4 * (x[1] - 1)
            + 2 * (x[0] - 3) ** 2 * (x[1] - 2)
    )
    

gamma = 0.009

# Посчитайте градиент и сделайте шаг в новую точку.

# Выполните расчёты для точки 0
grad = np.array([dx(x), dy(x)])
val_f = f(x)
one_step_print(0, x, val_f, grad)

grad = np.array([dx(x), dy(x)])
x = x - gamma*grad
val_f = f(x)
one_step_print(1, x, val_f, grad)

grad = np.array([dx(x), dy(x)])
x = x - gamma*grad
val_f = f(x)
one_step_print(2, x, val_f, grad)


grad = np.array([dx(x), dy(x)])
x = x - gamma*grad
val_f = f(x)
one_step_print(3, x, val_f, grad)

result = x
print(f(result))
```

**Критерий остановки**

```python
# Используем векторную запись точки. 
x = np.array([1, 2])

def f(x): 
	# f(x) = 0.5x^2 + 0.2y^2
	return 0.5*x[0]**2 + 0.2*x[1]**2 

def grad(x):
	# grad(x) = [x, 0.4y]
	dx = x[0]
	dy = 0.4 * x[1]
	return np.array([dx, dy])

# Коэффициент скорости спуска.
gamma = 1

# Наименьшее изменение значения функции, после которого алгоритм останавливается.
eps = 1e-3

# Наибольшее количество итераций, после которых цикл останавливается.
max_iters = 100

# Сохраняем старое значение функции.
f_old = f(x)
print("В начальной точке: f(x_0) =", f(x))

# Двигаем точку против градиента.
x = x - grad(x) * gamma

# Считаем новое значение функции.
f_new = f(x)

# Номер итерации.
i = 1

# Проверяем условия цикла.
while abs(f_new - f_old) > eps and i < max_iters:

	# Буква f перед строкой позволяет вставлять в неё переменные.
	print(f"f(x_{i}) =", f_new) 

	# Сохраняем старое значение функции.
	f_old = f_new

	# Двигаем точку против градиента.
	x = x - grad(x) * gamma

	# Считаем новое значение функции.
	f_new = f(x)
	
	# Увеличиваем номер итерации.
	i += 1

print("Успех, алгоритм завершил работу в точке:", x)
print("В конечной точке: f(x):", f(x))
```
